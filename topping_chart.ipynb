```python
from ipywidgets import widgets
from IPython.display import display, HTML, clear_output
from collections import Counter
import math

topping_chart = {
    'Raspberry': ['ATK', 9.0],
    'Apple Jelly': ['CRIT', 9.0],
    'Chocolate': ['CD', 3.0],
    'Almond': ['DMG RESIST', 4.1],
}

substat_chart = {
    'ATK': [1.0, 3.0],
    'DEF': [1.0, 3.0],
    'HP': [1.0, 3.0],
    'ATK SPD': [1.0, 3.0],
    'CRIT': [1.0, 3.0],
    'CD': [1.0, 2.0],
    'DMG RESIST': [1.0, 6.0],
    'CRIT RESIST': [3.0, 4.0],
    'AMPLIFY BUFF': [1.0, 2.0],
    'DEBUFF RESIST': [1.0, 2.0],
}

# Create widgets for main toppings
topping1 = widgets.Dropdown(
    options= list(topping_chart.keys()),
    value= list(topping_chart.keys())[0],
    description='Topping 1:'
)

topping2 = widgets.Dropdown(
    options= list(topping_chart.keys()),
    value= list(topping_chart.keys())[0],
    description='Topping 2:'
)

topping3 = widgets.Dropdown(
    options= list(topping_chart.keys()),
    value= list(topping_chart.keys())[0],
    description='Topping 3:'
)

topping4 = widgets.Dropdown(
    options= list(topping_chart.keys()),
    value= list(topping_chart.keys())[0],
    description='Topping 4:'
)

topping5 = widgets.Dropdown(
    options= list(topping_chart.keys()),
    value= list(topping_chart.keys())[0],
    description='Topping 5:'
)

substat_names = list(substat_chart.keys())

# Create widgets for substats
atk_numeric = widgets.FloatText(
    value=0,
    description='ATK%:'
)
atkspd_numeric = widgets.FloatText(
    value=0,
    description='ATKSPD%:'
)

cd_numeric = widgets.FloatText(
    value=0.0,
    description='CD%:'
)

crit_numeric = widgets.FloatText(
    value=0.0,
    description='CRIT%:'
)

dmgresist_numeric = widgets.FloatText(
    value=0.0,
    description='DR%:'
)


# Create a button
button = widgets.Button(description="Run Code")

# For output display
out = widgets.Output()

# Function to execute when the button is clicked
def on_button_click(b):

    # Access the selected values from the dropdowns
    selected_value_dropdown1 = topping1.value
    selected_value_dropdown2 = topping2.value
    selected_value_dropdown3 = topping3.value
    selected_value_dropdown4 = topping4.value
    selected_value_dropdown5 = topping5.value

    selected_toppings = [topping1.value, topping2.value, topping3.value, topping4.value, topping5.value]
    topping_counts = Counter(selected_toppings)

    wanted_atk = atk_numeric.value
    wanted_atkspd = atkspd_numeric.value
    wanted_cd = cd_numeric.value
    wanted_crit = crit_numeric.value
    wanted_dr = dmgresist_numeric.value

    wanted_stats_chart = {
        'ATK': wanted_atk,
        'ATK SPD': wanted_atkspd,
        'CRIT': wanted_crit,
        'CD': wanted_cd,
        'DMG RESIST': wanted_dr,
    }

    needed_stats_chart = {
        'ATK': wanted_atk,
        'ATK SPD': wanted_atkspd,
        'CRIT': wanted_crit,
        'CD': wanted_cd,
        'DMG RESIST': wanted_dr,
    }

    # use the topping types to determine how much substats you need
    for topping in selected_toppings:
        topping_stats = topping_chart[topping]
        topping_type = topping_stats[0]
        topping_value = topping_stats[1]
        needed_stats_chart[topping_type] = needed_stats_chart[topping_type]- topping_value

    with out:
        clear_output(wait=True)
        # Print the summary of how to get the percentages needed
        print('This is for:')
        for type, count in topping_counts.items():
            print(f"{count} {type} topping(s) ")

        print()
        for stat, value in wanted_stats_chart.items():
            if value != 0.0:
                print('You want ' + str(value) + '% ' + stat)
                needed_value = needed_stats_chart[stat]
                if needed_value < 0:
                    print(stat + '% is entirely gotten from main topping type')
                    print()
                elif needed_value > 0:
                    print(stat + ' has percentages between ' + str(substat_chart[stat][0]) + '% and ' + str(substat_chart[stat][1]) + '%')
                    if (needed_value/5) <= substat_chart[stat][1]:
                        print('You would need ' + str(needed_value / 5) + '% on all 5 toppings')
                        print('If you have the maximum value for the topping you would need ' +  str(math.ceil(needed_value / substat_chart[stat][1])) + ' toppings with the substat')
                        print()
                    else:
                        print('It is not possible')
                        print()
        print('--------------------')
    # Continue with the rest of your code based on the selected dropdown values

# Attach the function to the button's click event
button.on_click(on_button_click)

# Arrange the widgets into categories using VBox and HBox
dropdowns = widgets.HBox([topping1, topping2, topping3, topping4, topping5])
numerics = widgets.HBox([atk_numeric, atkspd_numeric, cd_numeric, crit_numeric, dmgresist_numeric])
widget_display = widgets.VBox([dropdowns, numerics, button, out])

# Display the widgets
display(widget_display)
```
